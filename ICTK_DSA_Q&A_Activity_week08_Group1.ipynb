# 2. Check whether female and male students scored the same marks.


```python
data.columns
```

```python
male = data[data['Gender'] == 'male']
male 
```

```python
female = data[data['Gender']=='female']
female 
```

```python
from scipy.stats import ttest_ind
t_value, p_value = ttest_ind(male['Total Marks'],female['Total Marks'])
print('t_value = %.4f, p_value = %.4f'% (t_value, p_value))
print('\nNull hypothesis: There is no difference between the marks scored by Male and Female students. \n \nAlternative_hypothesis: There is a difference between the marks scored by male and female students.')
if p_value <= 0.05:
  print('\nThe null hypothesis is rejected.\n There is a difference between the marks scored by Male and Female students')
else:
  print('\nWe fail to reject the null hypothesis.\nThere is no difference between the marks scored by Male and Female students')
    
```

# 3. Whether test preparation helps the students?

```python
test_prep=data[data['Test preparation']=='completed']
test_prep=test_prep.drop(['Test preparation','Gender'],axis=1)
no_test_prep=data[data['Test preparation']=='none']
no_test_prep=no_test_prep.drop(['Test preparation','Gender'],axis=1)
test_prep
```

```python
no_test_prep
```

```python
tvalue, pvalue = stats.ttest_ind(test_prep,no_test_prep)
print('tvalue:',tvalue,'pvalue:',pvalue)
```

```python
ttest_value=stats.t.ppf((1-0.05), 26)
ttest_value
```

```python
if (tvalue>ttest_value and pvalue <0.05):
    print('Reject Null Hypothesis: Test Preparation help students')
else:
    print('Reject Alternate Hypothesis:Test Preparation does not help students')
```
